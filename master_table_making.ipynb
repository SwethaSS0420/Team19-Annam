```python
import pandas as pd
import numpy as np

df = pd.read_csv(r"C:\Users\sweth\Downloads\master.csv")
df['crop'] = df['crop'].astype(str).str.strip().str.lower()

real_crop_ranges = {
    "rice": ("Alluvial, Loamy", (5.0, 6.5), (1000, 2500)),
    "maize": ("Loamy, Alluvial", (5.5, 7.0), (500, 800)),
    "chickpea": ("Loamy, Black", (6.0, 7.5), (600, 1000)),
    "kidneybeans": ("Loamy, Alluvial", (6.0, 7.5), (600, 1200)),
    "pigeonpeas": ("Black, Loamy", (5.0, 7.0), (600, 1000)),
    "mothbeans": ("Sandy, Red", (6.0, 7.5), (300, 500)),
    "mungbean": ("Loamy, Alluvial", (6.0, 7.5), (500, 1000)),
    "blackgram": ("Loamy, Alluvial", (6.0, 7.5), (500, 1000)),
    "lentil": ("Loamy, Alluvial", (6.0, 7.5), (500, 1000)),
    "pomegranate": ("Sandy, Red", (6.5, 7.5), (500, 800)),
    "banana": ("Alluvial, Loamy", (6.0, 7.5), (1000, 2000)),
    "mango": ("Alluvial, Loamy", (5.5, 7.5), (750, 2500)),
    "grapes": ("Sandy, Red", (6.5, 7.5), (500, 700)),
    "watermelon": ("Sandy, Loamy", (6.0, 6.8), (400, 600)),
    "muskmelon": ("Sandy, Loamy", (6.0, 6.8), (400, 600)),
    "apple": ("Loamy, Alluvial", (5.5, 6.5), (1000, 1250)),
    "orange": ("Loamy, Red", (6.0, 7.5), (1000, 2000)),
    "papaya": ("Sandy, Loamy", (6.0, 6.5), (1000, 1500)),
    "coconut": ("Sandy, Loamy", (5.5, 7.0), (1000, 2250)),
    "cotton": ("Black, Red", (5.8, 8.0), (500, 1200)),
    "jute": ("Alluvial, Loamy", (6.0, 7.5), (1500, 2000)),
    "coffee": ("Red, Loamy", (6.0, 6.5), (1500, 2500)),
    "sugarcane": ("Alluvial, Loamy", (6.0, 7.5), (1100, 1500)),
    "tobacco": ("Red, Sandy", (5.5, 6.5), (500, 1000)),
    "barley": ("Loamy, Alluvial", (6.0, 7.5), (400, 800)),
    "wheat": ("Loamy, Alluvial", (6.0, 7.0), (500, 1000)),
    "millets": ("Red, Sandy", (5.0, 6.5), (400, 800)),
    "oil seeds": ("Red, Sandy", (5.5, 7.5), (400, 1000)),
    "pulses": ("Red, Loamy", (6.0, 7.5), (400, 1000)),
    "ground nuts": ("Red, Sandy", (5.3, 6.3), (500, 1000))
}

print("Crop Match Counts:")
for crop in real_crop_ranges:
    count = df[df['crop'] == crop].shape[0]
    print(f" - {crop}: {count} rows")

for crop, (soil, ph_range, rain_range) in real_crop_ranges.items():
    mask = df['crop'] == crop
    df.loc[mask & df['soil_type'].isnull(), 'soil_type'] = soil
    df.loc[mask & df['ph'].isnull(), 'ph'] = np.round(np.random.uniform(*ph_range), 2)
    df.loc[mask & df['rainfall'].isnull(), 'rainfall'] = int(np.random.uniform(*rain_range))

df.to_csv(r"C:\Users\sweth\Downloads\master_filled.csv", index=False)
print("\nFilling complete. File saved as 'master_filled.csv'.")
```
